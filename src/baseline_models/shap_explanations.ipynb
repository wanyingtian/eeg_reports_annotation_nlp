```python
import torch
import shap
from shap import Explainer, plots
import joblib
import numpy as np
import os
from transformers import AutoModel, AutoTokenizer
import matplotlib.pyplot as plt

# Constants
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_DIR = 'trained_models'
result_columns = ["Focal Epi", "Gen Epi", "Focal Non-epi", "Gen Non-epi", "Abnormality"]
model_name = "bert_base"
MODEL_MAPPING = {
    "clinical-bert": "emilyalsentzer/Bio_ClinicalBERT",
    "bert_base": "bert-base-uncased",
    "bag_of_words": "BagOfWords" 
}
# Load BERT model and tokenizer
model = AutoModel.from_pretrained(MODEL_MAPPING[model_name]).to(device)
tokenizer = AutoTokenizer.from_pretrained(MODEL_MAPPING[model_name])

# Load trained logistic regression models
models = {}
for column in result_columns:
    model_path = os.path.join(MODEL_DIR, f"{column}_{model_name}_model.pkl")
    if os.path.exists(model_path):
        models[column] = joblib.load(model_path)
    else:
        print(f"Warning: Model for {column} not found!")

# Fix: Ensure the input is always a list of strings
def extract_embeddings(texts):
    """Extract CLS token embeddings from BERT for multiple inputs."""
    
    # Ensure `texts` is a list of strings (fixes SHAP modifying input format)
    if isinstance(texts, str):  
        texts = [texts]  # Convert single string to list
    elif isinstance(texts, np.ndarray):  
        texts = texts.tolist()  # Convert numpy array to list

    # Process text in batch
    with torch.no_grad():
        inputs = tokenizer(texts, return_tensors="pt", truncation=True, padding=True, max_length=512).to(device)
        outputs = model(**inputs)
        cls_embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()  # Extract CLS token embeddings
    
    return cls_embeddings  # Shape: (N, embedding_dim)



# Example input
text_simple = "This EEG is abnormal. IT has focal non-epliptiform spikes."
text = """
***FINAL REPORT***
 
                             DIAGNOSTIC REPORT - EEG
 
The patient is on 6-North with fluctuating level of consciousness.  Rule out
seizures.
 
DESCRIPTION OF THE RECORD
This EEG was captured from the patient who appeared asleep at times and, at
other times, appeared to be awake.
 
The EEG was generally represented by polymorphic delta and theta range
frequencies, low to moderate in amplitude, with no clear lateralized changes.
There were no epileptiform abnormalities.
 
Photic stimulation suggested some photic driving.
 
ECG tracing was abnormal.
 
IMPRESSION
This EEG is abnormal.  The abnormalities are moderate in degree and represented
by polymorphic slowing, consistent with nonspecific encephalopathy.

"""

for column in result_columns:
    print(f"{column}:")

    # Prediction function for SHAP
    def shap_prediction_function(texts):
        """Extract embeddings and return probabilities for multiple inputs."""
        embeddings = extract_embeddings(texts)  # Extract embeddings for all masked versions of the input
        #print("extracted embeddings")
        return models[column].predict_proba(embeddings)[:, 1]  # Prob of class 1

    # Initialize SHAP explainer
    explainer = Explainer(shap_prediction_function, tokenizer)

    # Compute SHAP values
    shap_values = explainer([text])
    print(shap_values)

        # 🔹 Print out final probability and decision
    pred_prob = models[column].predict_proba(extract_embeddings([text]))[:, 1][0]
    decision = "YES" if pred_prob > 0.5 else "NO"
    
    print(f"Prediction for {column}: {pred_prob:.3f} → Classified as: {decision}")

    plots.text(shap_values)
    shap.plots.bar(shap_values[0])
    #shap.summary_plot(shap_values)

```

```python
import shap
import joblib
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Constants
MODEL_DIR = 'trained_models'
result_columns = ["Focal Epi", "Gen Epi", "Focal Non-epi", "Gen Non-epi", "Abnormality"]
model_name = "bag_of_words"

# Load trained logistic regression models
models = {}
for column in result_columns:
    model_path = os.path.join(MODEL_DIR, f"{column}_{model_name}_model.pkl")
    if os.path.exists(model_path):
        models[column] = joblib.load(model_path)
    else:
        print(f"Warning: Model for {column} not found!")

# Load vectorizer
vectorizer_path = os.path.join(MODEL_DIR, "bag_of_words_vectorizer.pkl")
if os.path.exists(vectorizer_path):
    vectorizer = joblib.load(vectorizer_path)
    feature_names = vectorizer.get_feature_names_out()
else:
    raise FileNotFoundError("Vectorizer not found! Ensure it was saved during training.")



# Example input text
text = """
***FINAL REPORT***
 
                             DIAGNOSTIC REPORT - EEG
 
The patient is on 6-North with fluctuating level of consciousness.  Rule out
seizures.
 
DESCRIPTION OF THE RECORD
This EEG was captured from the patient who appeared asleep at times and, at
other times, appeared to be awake.
 
The EEG was generally represented by polymorphic delta and theta range
frequencies, low to moderate in amplitude, with no clear lateralized changes.
There were no epileptiform abnormalities.
 
Photic stimulation suggested some photic driving.
 
ECG tracing was abnormal.
 
IMPRESSION
This EEG is abnormal.  The abnormalities are moderate in degree and represented
by polymorphic slowing, consistent with nonspecific encephalopathy.

"""

# Convert text to BoW vector
bow_vector = vectorizer.transform([text]).toarray()  # Transform input text into sparse matrix
decisions = {}
# Apply SHAP to each classifier
for column in result_columns:
    print(f"\n🔹 {column} SHAP Analysis:")

    if column not in models:
        print(f"Skipping {column} due to missing model.")
        continue

    # Get the trained logistic regression model
    model = models[column]

    background = np.zeros((1, bow_vector.shape[1]))  # Zero-vector baseline
    explainer = shap.Explainer(model, background)
    shap_values = explainer(bow_vector)

    # Print Prediction
    pred_prob = model.predict_proba(bow_vector)[0, 1]  # Probability of class 1
    decision = "YES" if pred_prob > 0.5 else "NO"

    print(f"Prediction for {column}: {pred_prob:.3f} → Classified as: {decision}")
    decisions[column] = decision


    coefficients = model.coef_.flatten()
    #print("coefficients:",coefficients)
    word_contributions = [(feature_names[i], bow_vector.flatten()[i] * coefficients[i]) for i in range(len(coefficients))]
    word_contributions = sorted(word_contributions, key=lambda x: abs(x[1]), reverse=True)[:30]

    # Print the top words and their contributions
    print(f"1. Top contributing words by coefficients for {column}:")
    for word, weight in word_contributions:
        print(f"  - {word}: {weight:.4f}")

    # Display top words influencing the decision
# Display SHAP values for tokens with non-zero contribution
    print(f"\n2. SHAP Values for Tokens Present in Input for {column}:")
    shap_contribs = list(zip(feature_names, shap_values.values.flatten()))
    nonzero_contribs = [(w, v) for w, v in shap_contribs if v != 0]
    top_contribs = sorted(nonzero_contribs, key=lambda x: abs(x[1]), reverse=True)[:20]

    for word, val in top_contribs:
        print(f"{word:25} {val:+.4f}")


    # Get the absolute SHAP values for feature importance
    # shap_importance = np.abs(shap_values.values).flatten()

    # Sort words by their SHAP importance
    # top_n = 30  # Number of top words to display
    # top_shap_words = sorted(
    #     zip(feature_names, shap_importance), key=lambda x: x[1], reverse=True
    # )[:top_n]

    # Print the top SHAP words
    # print("\n🔍 Top SHAP Contributing Words:")
    # for word, shap_value in top_shap_words:
    #     print(f"  - {word}: {shap_value:.4f}")

    # # Plot bar chart of SHAP contributions
    # plt.figure(figsize=(10, 6))
    # words, shap_scores = zip(*top_shap_words)
    # plt.barh(words[::-1], shap_scores[::-1])  # Reverse for best visualization
    # plt.xlabel("SHAP Value (Impact on Model Output)")
    # plt.ylabel("Top Words")
    # plt.title("Top 30 Words Contributing to Prediction")
    # plt.show()


    # Visualize SHAP explanation
    shap.summary_plot(shap_values, feature_names=feature_names)
    #shap.plots.bar(shap_values[0])
    #shap.plots.bar(shap_values)
    #shap.plots.text(shap_values)

print(decisions)
```
